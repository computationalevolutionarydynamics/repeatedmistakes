```python
%matplotlib inline
import pandas as pd
import seaborn as sb

# Grab data from file and sort
simulation_data = pd.DataFrame.from_csv("simulation_results.txt", index_col=[0,1])
simulation_data.sort_index(inplace=True)

calculation_data = pd.read_csv("calculation_results_highaccuracy.txt", index_col=[1,2])
calculation_data.sort_index(inplace=True)

# Subtract the simulations from the calculations, then divide to get relative error
sub = calculation_data.sub(simulation_data[['TFNT','InverseTitForTat']])
div = sub.div(simulation_data[['TFNT', 'InverseTitForTat']])

TFNT = div['TFNT'].unstack()
#sb.heatmap(TFNT, annot=True)

sb_plot = sb.heatmap(TFNT,annot=True,square=True)
```
